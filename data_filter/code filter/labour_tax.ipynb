```python
import pandas as pd
```

```python
tax = pd.read_csv(r"C:/Downloads/new_data/thuế lao động.csv")
eu_regions = pd.read_csv(r"C:/Downloads/new_data/eu_country.csv")
tax.head()
```

```python
tax.columns
```

```python
cols_keep = [
    "TIME_PERIOD",          
    "REF_AREA",             
    "Zone de référence",
    "Type de ménage",
    "Mesure",      
    "INCOME_PRINCIPAL", 
    "INCOME_SPOUSE",            
    "OBS_VALUE",   
    "UNIT_MEASURE",         
    # "UNIT_MULT",
]
```

```python
tax[cols_keep]
```

```python
tax["Type de ménage"].unique()
```

```python
# 1. Lọc các cột đã chọn
test = tax[cols_keep]

# 2. Đổi tên cột cho gọn & chuẩn
test = test.rename(columns={
    "TIME_PERIOD": "Year",
    "Zone de référence": "Country",
    "Mesure" : "Measure",
    "Type de ménage":"Household_type",
    "OBS_VALUE": "Value_tax",
    # "UNIT_MULT": "Unit_mult",
    "UNIT_MEASURE" : "Unit_measure"
})

```

```python
eu_country = test["REF_AREA"].isin(eu_regions["REF_AREA"])
test = test[eu_country]
test
```

```python
test["Measure"].unique()
```

```python
measure_map = {
    # "Revenus bruts avant impôts": "income_before_tax",
    # "Revenus nets après impôts en monnaie nationale": "income_after_tax"
    "Taux moyen net d'imposition à l'impôt sur le revenu": "net_average_tax_rate"

}

test["Measure"]= test["Measure"].replace(measure_map)
test = test[test["Measure"].isin(measure_map.values())]
test = test[test["Unit_measure"] == "PT_WG_EARN_G"]
test = test.rename(
    columns={"Value_tax": "Net_average_tax_rate_pct"}
)

```

```python
test["INCOME_PRINCIPAL"].unique()
```

```python
fra = test[(test["REF_AREA"] == "FRA") & (test["Year"] == 2024)]
fra
```

```python
income = test["INCOME_PRINCIPAL"] != "AW167"
test = test[income]
# No help
no_help = test["Net_average_tax_rate_pct"] > 0
test = test[no_help]

def convert(value) :
    return int(value[2:]) if value.startswith("AW") else 0

test["INCOME_SPOUSE"] = test["INCOME_SPOUSE"].apply(convert)
test["INCOME_PRINCIPAL"] = test["INCOME_PRINCIPAL"].apply(convert)


test["Net_average_tax_rate_pct"] = test["Net_average_tax_rate_pct"].round(2)
test[(test["REF_AREA"] == "FRA") & (test["Year"] == 2024)]
```

```python
test = test.dropna(subset = ["Net_average_tax_rate_pct"])
test
```

```python
test["Unit_measure"].unique()
test["Unit_measure"] = test["Unit_measure"].replace("PT_WG_EARN_G","PT")
```

```python
couple = (test["INCOME_PRINCIPAL"] == 100) & (test["INCOME_SPOUSE"] == 67)
single = (test["INCOME_PRINCIPAL"] == 100) & (test["INCOME_SPOUSE"] == 0)
check = test[couple | single]

check.groupby("Year")[["Country","Household_type","Measure","INCOME_PRINCIPAL","INCOME_SPOUSE"]].nunique()
```

```python
test = test.reset_index(drop = True)
test
```
