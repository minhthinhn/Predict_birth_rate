```python
import pandas as pd
```

```python
home_spending = pd.read_csv(r"C:/Downloads/new_data/Chi tiêu tiêu dùng cuối cùng hàng năm của hộ gia đình theo mục đích (COICOP).csv")
eu_regions = pd.read_csv(r"C:/Downloads/new_data/eu_country.csv")
home_spending.head()
```

```python
home_spending.columns
```

```python
cols_keep = [
    "TIME_PERIOD",          # năm
    "REF_AREA",             # mã quốc gia
    "Zone de référence",    # tên quốc gia      
    "Dépense",              #Tên đầy đủ của loại chi tiêu    
    "OBS_VALUE",             
    "UNIT_MULT",
    "CURRENCY"
]
```

```python
home_spending[cols_keep]
```

```python
# 1. Lọc các cột bạn đã chọn
test = home_spending[cols_keep]

# 2. Đổi tên cột cho gọn & chuẩn
test = test.rename(columns={
    "TIME_PERIOD": "Year",
    "Zone de référence": "Country",
    "Dépense": "Expenditure_name",
    "OBS_VALUE": "value_raw",
    "UNIT_MULT": "Unit_mult",
    "CURRENCY": "Currency"
})

```

```python
test["Expenditure_name"].unique()
```

```python
expenditure_map = {
    "Produits alimentaires et boissons non alcoolisées": "Food",
    "Loisirs et culture": "Recreation_Culture",
    "Meubles, articles de ménage et entretien courant du foyer": "Household_Goods",
    "Santé": "Health",
    "Communications": "Communication",
    "Enseignement": "Education",
    "Boissons alcoolisées, tabac et stupéfiants": "Alcohol_Tobacco",
    "Articles d'habillement et chaussures": "Clothing_Footwear",
    "Biens et services divers": "Other_Services",
    "Logement, eau, gaz, électricité et autres combustibles": "Housing_Utilities",
    "Transports": "Transport",
    "Restaurants et hôtels": "Restaurants_Hotels"
}


test["Expenditure_name"] = test["Expenditure_name"].replace(expenditure_map)
```

```python
test["Expenditure_name"].unique()
```

```python
total_df = (
    test[test["Expenditure_name"] == "Total"]
    [["Country", "Year", "value_raw"]]
    .rename(columns={"value_raw": "total_expenditure"})
)
detail_df = test[test["Expenditure_name"] != "Total"].copy()

detail_df = detail_df.merge(
    total_df,
    on=["Country", "Year"],
    how="left"
)
detail_df["Expenditure_Share %"] = (
    detail_df["value_raw"] / detail_df["total_expenditure"] * 100
)

```

```python
detail_df["Expenditure_Share %"] = detail_df["Expenditure_Share %"].round(2)
detail_df
```

```python
test = detail_df.drop(columns= ["total_expenditure"])
```

```python
test = test.dropna(subset= ["value_raw","Expenditure_Share %"])
```

```python
eu =test["REF_AREA"].isin(eu_regions["REF_AREA"])
test = test[eu]
```

```python
test
```

```python
france = (test["REF_AREA"] == "FRA") & (test["Year"] == 1995)
test[france]
```

```python
# Số quốc gia và số nhóm chi tiêu chuẩn
N_country = test["Country"].nunique()
N_expenditure = test["Expenditure_name"].nunique()
```

```python
countries_per_year = (
    test
    .groupby("Year")["Country"]
    .nunique()
    .reset_index(name="n_country")
)

countries_per_year["country_complete"] = (
    countries_per_year["n_country"] == N_country
)
```

```python
expenditures_per_year = (
    test
    .groupby("Year")["Expenditure_name"]
    .nunique()
    .reset_index(name="n_expenditure")
)

expenditures_per_year["expenditure_complete"] = (
    expenditures_per_year["n_expenditure"] == N_expenditure
)
```

```python
year_completeness = (
    countries_per_year
    .merge(expenditures_per_year, on="Year")
)
year_completeness
```

```python
year = test["Year"] != "2023"
test = test[year]
test = test.reset_index(drop= True)
```
