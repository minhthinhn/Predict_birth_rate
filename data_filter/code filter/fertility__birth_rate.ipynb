```python
cols_keep = [
    "TIME_PERIOD",
    "REF_AREA",          # code area
    "Reference area",    # name area
    "COUNTRY",           # code country
    "Country",           # name country
    "TERRITORIAL_LEVEL",
    "MEASURE",
    "UNIT_MEASURE",
    "Unit of measure",
    "AGE",
    "SEX",
    "OBS_VALUE"          # target value
]
```

```python
fertility_eu = fertility[cols_keep].copy()
fertility_eu
```

```python

cols_name = {
    "TIME_PERIOD" : "Year",
    "REF_AREA" : "REF_REGION",
    "Zone de référence" : "Area",
    "AGE" : "Age",
    "SEX" : "Sex",
    "COUNTRY":"REF_AREA",
    "Pays" : "Country",
    "TERRITORIAL_LEVEL" : "Level",
    "Unité de mesure" : "Unit_measure",
    "MEASURE" :"Measure",
    "OBS_VALUE" : "Value"
}
```

```python
fertility_eu = fertility_eu.rename(columns= cols_name)
fertility_eu
```

```python
#code filter
year = (fertility_eu["Year"] == 2022)
fertility_eu = fertility_eu[year]
fertility_eu = fertility_eu[fertility_eu["REF_AREA"].isin(eu["REF_AREA"])]

# Fertility and birth rate (level Country and Area)
fertility_area = fertility_eu.copy()
birth_rate_condition = (fertility_area["Measure"] == "LIVE_BIRTHS") & (fertility_area["Age"] == "Y0") & (fertility_area["Sex"] == "_T") & (fertility_area["Unit_measure"] == "Pour 1 000 habitants")
fertility_condition = (fertility_area["Measure"] == "FERT_RATIO") & (fertility_area["Age"] == "_T")
fertility_area = fertility_area[fertility_condition | birth_rate_condition]

fertility_area.drop(columns= ["Age","Sex","Unit_measure"],inplace = True)
# birth rate


birth_age = fertility_eu["Age"] == "Y0"
total_sex = fertility_eu["Sex"] == "_T"
birth_rate_eu = fertility_eu[total_sex & birth_age]

# fertility
fertility_eu = fertility_eu[fertility_eu["Measure"] == "FERT_RATIO"]
total_age = fertility_eu["Age"] == "_T"
fertility_eu = fertility_eu[total_age]

```

```python
country = birth_rate_eu["Level"] == "CTRY"

birth_rate_eu = birth_rate_eu[country]
Per_1000_inhabitants = birth_rate_eu["Unit_measure"] == "Pour 1 000 habitants"
birth_rate_eu = birth_rate_eu[Per_1000_inhabitants]

birth_rate_eu.dropna(subset= "Value",inplace= True)
birth_rate_eu.drop(columns= ["Age","Sex","Unit_measure"],inplace = True)
```

```python
country = fertility_eu["Level"] == "CTRY"
fertility_eu = fertility_eu[country]
fertility_eu.dropna(subset= "Value",inplace= True)
fertility_eu.drop(columns= ["Age","Sex","Unit_measure"],inplace = True)
```
