```python
import pandas as pd
```

```python
employment = pd.read_csv(r"C:/Downloads/new_data/Tỷ lệ việc làm và thất nghiệp theo nhóm tuổi.csv")
eu_regions = pd.read_csv(r"C:/Downloads/new_data/eu_country.csv")
employment.head()
```

```python
employment.columns
```

```python
cols_keep = [
    "REF_AREA",
    "Reference area",
    "TIME_PERIOD",
    "Sex",
    "Age",
    "Measure",
    "OBS_VALUE",
]
```

```python
test = employment[cols_keep]
test.sort_values(axis = 0,by = ["TIME_PERIOD"])
```

```python
rename_map = {
    "Reference area": "Country",
    "TIME_PERIOD": "Year",
    "OBS_VALUE": "Value_emp_rate"
}

test = test.rename(columns=rename_map)
test
```

```python
eu_country = test["REF_AREA"].isin(eu_regions["REF_AREA"])
test = test[eu_country]
test
```

```python
test["Age"] = test["Age"].replace("De 20 à 64 ans","20-64")
test

```

```python
test = test.dropna(subset= ["Value_emp_rate"])
test["Value_emp_rate"] = test["Value_emp_rate"].round(2)
test
```

```python
test.isna().any(axis = 0).sum()
```

```python
test = test.reset_index(drop = True)
```
