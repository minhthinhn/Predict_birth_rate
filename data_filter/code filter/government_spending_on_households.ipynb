```python
import pandas as pd
```

```python
gor = pd.read_csv(r"C:/Downloads/new_data/Chi tiêu công cho gia đình theo loại hình chi tiêu (tiền mặt và hiện vật) tính theo % GDP.csv")
eu_regions = pd.read_csv(r"C:/Downloads/new_data/eu_country.csv")
gor.head()
```

```python
gor.columns
```

```python
cols_keep = [
    "TIME_PERIOD",          # năm
    "REF_AREA",             # mã quốc gia
    "Zone de référence",    # tên quốc gia
    "UNIT_MEASURE",
    "Type de programme",    # tên chính sách (UI OECD)
    "Type de dépenses",     # cash / in-kind
    "OBS_VALUE",             # % GDP
    "UNIT_MULT"
]
```

```python
gor[cols_keep]
```

```python
# 1. Lọc các cột bạn đã chọn
test = gor[cols_keep]

# 2. Đổi tên cột cho gọn & chuẩn
test = test.rename(columns={
    "TIME_PERIOD": "Year",
    "Zone de référence": "Country",
    "UNIT_MEASURE": "Unit_measure",
    "Type de programme": "Program",
    "Type de dépenses": "Spending_type",
    "OBS_VALUE": "Value of GPD %",
    "UNIT_MULT": "Unit_mult"
})

test
```

```python
test["Unit_measure"].unique()
```

```python
test["Unit_measure"] = test["Unit_measure"].replace("PT_B1GQ",r"% GDP")
gpd = test["Unit_measure"] == r"% GDP"
test = test[gpd]
test
```

```python
test["Program"].unique()
```

```python
program_map = {
    "Famille": "family_total",
    "Allocations familiales": "family_allowances",
    "Congé de maternité et congé parental": "parental_leave",
    "Aide à domicile / Hébergement": "home_support",
    "Education et à l'accueil des jeunes enfants (EAJE)": "childcare_education"
}
test["Program"] = test["Program"].map(program_map)
test = test.dropna(subset = ["Value of GPD %"])
```

```python
test
```

```python
test["Spending_type"].unique()
```

```python
spending_type_map = {
    "Dépense en nature": "in_kind",
    "Dépense en espèces": "cash",
    "Total" : "Total"
}

test["Spending_type"] = test["Spending_type"].map(spending_type_map)
```

```python
country = test["REF_AREA"].isin(eu_regions["REF_AREA"])
year = test["Year"] < 2022
test = test[country & year]
```

```python
test = test.reset_index(drop = True)
test
```

```python
value = test["Value of GPD %"] == 0
zeros = test[value].sort_values(by= ["Year"])
```
